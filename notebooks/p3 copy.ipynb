# Load data

```python
X_train = pd.read_csv('../data/data_numerical/multiple_classification/X_train_1.csv', index_col='Unnamed: 0')
X_test = pd.read_csv('../data/data_numerical/multiple_classification/X_test_1.csv', index_col='Unnamed: 0')
y_train = pd.read_csv('../data/data_numerical/multiple_classification/y_train_1.csv', index_col='Unnamed: 0')
y_train.columns = [''] * len(y_train.columns)
y_test = pd.read_csv('../data/data_numerical/multiple_classification/y_test_1.csv', index_col='Unnamed: 0')
y_test.columns = [''] * len(y_test.columns)
```

```python
X_train.select_dtypes(include='number').columns
```

```python
X_train = X_train.drop([ 
    'attack_cat_normal'], axis=1)
X_test = X_test.drop([ 
    'attack_cat_normal'], axis=1)
```

```python
print("train data shape", X_train.shape, y_train.shape)
print("test data shape", X_test.shape, y_test.shape)
```

# Machine Learning model

```python
import os
import pandas as pd
import numpy as np

from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

# Logistic Regression Model

```python
prams={
    'alpha':[10 ** x for x in range(-4, 1)],
     'max_iter':[5, 10, 20, 50, 100],
    'eta0': [10 ** x for x in range(-4, 1)]
}
lr_cfl=GridSearchCV(SGDClassifier(penalty='l2', loss='log', n_jobs = -1), param_grid=prams,verbose=10,n_jobs=-1)
lr_cfl.fit(X_train,y_train)
```

```python
print(lr_cfl.best_params_)
```

```python
logisticR=SGDClassifier(alpha=lr_cfl.best_params_['alpha'],eta0=lr_cfl.best_params_['eta0'], penalty='l2', loss='log', n_jobs = -1, max_iter=lr_cfl.best_params_['max_iter'])
logisticR.fit(X_train,y_train)
```

```python
logisticR_acc = (logisticR.score(X_test, y_test)) * 100
print("ACC: ", logisticR_acc, '%')
```

# Support Vector Machine Model

```python
prams={
    'alpha':[10 ** x for x in range(-4, 1)],
     'max_iter':[5, 10, 20, 50, 100],
    'eta0': [10 ** x for x in range(-4, 1)]
}
svm_cfl=GridSearchCV(SGDClassifier(penalty='l2', loss='hinge', n_jobs = -1), param_grid=prams,verbose=10,n_jobs=-1)
svm_cfl.fit(X_train,y_train)
```

```python
print(svm_cfl.best_params_)
```

```python
svm=SGDClassifier(alpha=svm_cfl.best_params_['alpha'],eta0=svm_cfl.best_params_['eta0'], penalty='l2', loss='hinge', n_jobs = -1, max_iter=svm_cfl.best_params_['max_iter'])
svm.fit(X_train,y_train)
sig_clf_svm = CalibratedClassifierCV(svm, method="sigmoid")
sig_clf_svm.fit(X_train, y_train)
```

```python
sig_clf_svm_acc = sig_clf_svm.score(X_test, y_test) * 100
print("ACC: ", sig_clf_svm_acc, '%')
```

# Random Forest Model

```python
param_grid = {"n_estimators": [10,100,500,1000, 2000],
    "min_samples_split": [50, 80, 120, 200],
              "max_depth": [3, 5, 10, 50, 100]}
rfc = RandomForestClassifier(criterion='gini', random_state=42, n_jobs=-1)
gridCV_rfc = GridSearchCV(rfc, param_grid, cv=3, verbose=10, n_jobs=-1)
gridCV_rfc.fit(X_train, y_train)
```

```python
print(gridCV_rfc.best_params_)
```

```python
rfc= RandomForestClassifier(criterion='gini', random_state=42, n_jobs=-1, max_depth=gridCV_rfc.best_params_['max_depth'],min_samples_split=gridCV_rfc.best_params_['min_samples_split'], n_estimators=gridCV_rfc.best_params_['n_estimators'])
rfc.fit(X_train,y_train)
sig_clf_rfc = CalibratedClassifierCV(rfc, method="sigmoid")
sig_clf_rfc.fit(X_train, y_train)
```

```python
sig_clf_rfc_acc = (sig_clf_rfc.score(X_test, y_test))*100
print("ACC: ", sig_clf_rfc_acc, '%')
```

# Stacking classifier

```python
clf1 = SGDClassifier(alpha=0.0001,eta0=1, penalty='l2', loss='log', n_jobs = -1, max_iter=10)
clf1.fit(X_train, y_train)
sig_clf1 = CalibratedClassifierCV(clf1, method="sigmoid")

clf2 = SGDClassifier(alpha=0.0001,eta0=0.0001, penalty='l2', loss='hinge', n_jobs = -1, max_iter=5)
clf2.fit(X_train, y_train)
sig_clf2 = CalibratedClassifierCV(clf2, method="sigmoid")


clf3 = RandomForestClassifier(criterion='gini', random_state=42, n_jobs=-1, max_depth=50,min_samples_split=50, n_estimators=10)
clf3.fit(X_train, y_train)
sig_clf3 = CalibratedClassifierCV(clf3, method="sigmoid")
```

```python
alpha = [0.0001,0.001,0.01,0.1,1,10] 
best_alpha = 999
for i in alpha:
    lr = LogisticRegression(C=i)
    sclf = StackingClassifier(estimators=[("lr",sig_clf1), ("svm", sig_clf2),("RF", sig_clf3)], final_estimator=lr, n_jobs=-1)
    sclf.fit(X_train, y_train)
```

```python
lr = LogisticRegression(C=10)
sig_clf_sc = StackingClassifier(estimators=[("lr",sig_clf1), ("svm", sig_clf2),("RF", sig_clf3)], final_estimator=lr, n_jobs=-1)
sig_clf_sc.fit(X_train, y_train)
```

```python
sig_clf_sc_acc = sig_clf_sc.score(X_test, y_test)*100
print("ACC: ", sig_clf_sc_acc, '%')
```

# Hard Voting 

```python
from sklearn import datasets, linear_model, svm, neighbors
from sklearn.ensemble import VotingClassifier
```

```python
hard_learner_1 = neighbors.KNeighborsClassifier(n_neighbors=5)
hard_learner_2 = linear_model.Perceptron(tol=1e-2, random_state=0)
hard_learner_3 = svm.SVC(gamma=0.001)
```

```python
hard_voting = VotingClassifier([('KNN', hard_learner_1),
                            ('Prc', hard_learner_2),
                            ('SVM', hard_learner_3)])
```

```python
hard_voting.fit(X_train, y_train)
```

```python
hard_voting_acc = hard_voting.score(X_test, y_test) * 100
print('ACC: ', hard_voting_acc, '%')
```

# Sort Voting 

```python
from sklearn import naive_bayes, svm, neighbors
from sklearn.ensemble import VotingClassifier
```

```python
soft_learner_1 = neighbors.KNeighborsClassifier(n_neighbors=5)
soft_learner_2 = naive_bayes.GaussianNB()
soft_learner_3 = svm.SVC(gamma=0.001, probability=True)
```

```python
soft_voting = VotingClassifier([('KNN', soft_learner_1),
                                ('NB', soft_learner_2),
                                ('SVM', soft_learner_3)],
                                voting = 'soft')
```

```python
soft_voting.fit(X_train, y_train)
```

```python
soft_voting_acc = soft_voting.score(X_test, y_test) * 100
print('ACC: ', soft_voting_acc, '%')
```
